```python
def combinations(ind, arr, stack,ans):
    if ind == len(arr):
        ans.append(stack.copy())
        return 
    
    stack.append(arr[ind])
    combinations(ind + 1, arr, stack, ans)
    stack.pop()
    combinations(ind + 1, arr, stack, ans)
```
